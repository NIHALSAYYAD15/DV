Implementation of distribution-based graphs and chart Density, Histogram. 

### 1. Density 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data generation
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Density Plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data, fill=True)
plt.title('Density Plot')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

### 2. Histogram 

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data generation
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
