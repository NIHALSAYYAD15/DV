### 1. Stacked Bar :

```python
import matplotlib.pyplot as plt

categories = ['Category1', 'Category2', 'Category3']
labels = ['Label1', 'Label2', 'Label3']
data = {
    'Label1': [10, 15, 20],
    'Label2': [5, 12, 8],
    'Label3': [8, 10, 15]
}

# Plotting the stacked bar chart
fig, ax = plt.subplots()
bottom = [0] * len(categories)

for i, label in enumerate(labels):
    values = data[label]
    ax.bar(categories, values, label=label, bottom=bottom)
    bottom = [bottom[j] + values[j] for j in range(len(categories))]

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Simple Stacked Bar Chart')
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV
data = pd.read_csv('Data/data5.csv', index_col=0)

# Extract categories and labels
categories = data.index.tolist()
labels = data.columns.tolist()

# Plotting the stacked bar chart
fig, ax = plt.subplots()

# Initialize bottom values for stacking
bottom = [0] * len(categories)

for i, label in enumerate(labels):
    values = data[label].tolist()
    ax.bar(categories, values, label=label, bottom=bottom)
    bottom = [bottom[j] + values[j] for j in range(len(categories))]

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Simple Stacked Bar Chart')
ax.legend()

plt.show()
```

### 3. Pie

```python
import matplotlib.pyplot as plt

def pie_chart(data, labels, title):
    plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99'])
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.title(title)
    plt.show()

# Example usage:
data = [40, 30, 30]
labels = ['Label1', 'Label2', 'Label3']
title = 'Pie Chart Example'

pie_chart(data, labels, title)
```
