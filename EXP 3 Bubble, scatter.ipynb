Implementation of correlation - based graphs and charts â€“ Bubble, scatter.


### 1. Bubble:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a simple dataset
np.random.seed(0)
size = 30
x = np.random.rand(size)
y = 2 * x + 1 + 0.1 * np.random.randn(size)  # y is correlated with x
z = np.random.rand(size)  # z is another variable
```

```python
# Bubble Chart with Color
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=z * 100, alpha=0.7, edgecolors="grey", linewidth=0.5)
plt.title('Bubble Chart - Correlation between x, y, and bubble size')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Bubble Size')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV
data = pd.read_csv('Data/data3.csv')

# Extract columns for x, y, z (bubble size)
x = data['x']
y = data['y']
z = data['z']

# Plotting the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=z * 100, alpha=0.7, edgecolors="grey", linewidth=0.5)
plt.title('Bubble Chart - Correlation between x, y, and bubble size')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Bubble Size')
plt.show()
```

### 2. scatter:

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=z, alpha=0.7, edgecolors="grey", linewidth=0.5)
plt.title('Scatter Plot - Correlation between x, y, and color')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Color')
plt.show()
```
