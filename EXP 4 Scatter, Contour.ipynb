Implementation of correlation - based graphs and charts â€“ Scatter, Contour.

### 1. Scatter:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a simple dataset
np.random.seed(0)
size = 30
x = np.random.rand(size)
y = 2 * x + 1 + 0.1 * np.random.randn(size)  # y is correlated with x
z = np.random.rand(size)  # z is another variable
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=z, alpha=0.7, edgecolors="grey", linewidth=0.5)
plt.title('Scatter Plot - Correlation between x, y, and color')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Color')
plt.show()
```

### 2. Contour:

```python
plt.figure(figsize=(10, 6))
contour = plt.tricontourf(x, y, z, cmap='viridis')
plt.title('Contour Plot - Correlation between x, y, and contour levels')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(contour, label='Contour Levels')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
feature_x = np.linspace(-5.0, 3.0, 70) 
feature_y = np.linspace(-5.0, 3.0, 70) 
  
# Creating 2-D grid of features 
[X, Y] = np.meshgrid(feature_x, feature_y) 
  
fig, ax = plt.subplots(1, 1) 
  
Z = X ** 2 + Y ** 2
  
# plots filled contour plot 
ax.contourf(X, Y, Z) 
  
ax.set_title('Filled Contour Plot') 
ax.set_xlabel('feature_x') 
ax.set_ylabel('feature_y') 
  
plt.show() 
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.interpolate import griddata

# Load the data from CSV
data = pd.read_csv('Data/data4.csv')

# Extract feature_x, feature_y, and Z values
feature_x = data['feature_x']
feature_y = data['feature_y']
Z = data['Z']

# Create a grid of points
x_grid, y_grid = np.meshgrid(feature_x, feature_y)

# Interpolate Z to fit the grid
Z_grid = griddata((feature_x, feature_y), Z, (x_grid, y_grid), method='linear')

# Plot filled contour plot
fig, ax = plt.subplots(1, 1)
contour = ax.contourf(x_grid, y_grid, Z_grid)
ax.set_title('Filled Contour Plot')
ax.set_xlabel('feature_x')
ax.set_ylabel('feature_y')
plt.colorbar(contour, ax=ax, label='Z')

plt.show()
```
