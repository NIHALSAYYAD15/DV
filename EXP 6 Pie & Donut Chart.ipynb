### 1. Pie

```python
import matplotlib.pyplot as plt

def pie_chart(data, labels, title):
    plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99'])
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.title(title)
    plt.show()

# Example usage:
data = [40, 30, 30]
labels = ['Label1', 'Label2', 'Label3']
title = 'Pie Chart Example'

pie_chart(data, labels, title)
```

### 2. Donut Chart  :

```python
import matplotlib.pyplot as plt

# Data
labels = ['Part 1', 'Part 2', 'Part 3']
sizes = [30, 40, 30]

# Plot
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.6))

# Draw center circle to create donut chart
centre_circle = plt.Circle((0, 0), 0.1, color='white', linewidth=0)
ax.add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

plt.title('Donut Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV
data = pd.read_csv('Data/data6.csv')

# Extract data and labels
sizes = data['sizes']
labels = data['labels']

# Plotting the donut chart
fig, ax = plt.subplots()

# Plot the pie chart
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.6))

# Draw center circle to create a donut chart effect
centre_circle = plt.Circle((0, 0), 0.7, color='white', linewidth=0)
ax.add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

plt.title('Donut Chart')
plt.show()
```
