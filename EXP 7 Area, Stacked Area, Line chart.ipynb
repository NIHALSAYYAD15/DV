Implementation of data over time (Temporal) - based graphs and charts â€“ Area, Stacked Area, Line chart.


### 1. Area 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data generation
np.random.seed(0)
dates = pd.date_range('2024-01-01', periods=25)
data = pd.DataFrame(np.random.rand(25, 4), index=dates, columns=list('ABCD'))

# Area Chart
plt.figure(figsize=(10, 6))
data.plot.area()
plt.title('Area Chart')
plt.xlabel('Date')
plt.ylabel('Values')
plt.show()
```

### 2. Stacked Area  :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data generation
np.random.seed(0)
dates = pd.date_range('2024-01-01', periods=50)
data = pd.DataFrame(np.random.rand(50, 4), index=dates, columns=list('ABCD'))

# Stacked Area Chart
plt.figure(figsize=(10, 6))
data.plot.area(stacked=True)
plt.title('Stacked Area Chart')
plt.xlabel('Date')
plt.ylabel('Values')
plt.show()
```

### 3. Line chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data generation
np.random.seed(0)
dates = pd.date_range('2024-01-01', periods=50)
data = pd.DataFrame(np.random.randn(50, 4), index=dates, columns=list('ABCD'))

# Line Chart
plt.figure(figsize=(10, 6))
data.plot(kind='line')
plt.title('Line Chart')
plt.xlabel('Date')
plt.ylabel('Values')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
data = pd.read_csv('Data/data7.csv', index_col=0, parse_dates=True)

# Line Chart
plt.figure(figsize=(10, 6))
data.plot(kind='line', marker='o')  # Using marker='o' for better visibility of data points
plt.title('Line Chart')
plt.xlabel('Date')
plt.ylabel('Values')
plt.grid(True)  # Add grid lines for better readability
plt.show()
```
