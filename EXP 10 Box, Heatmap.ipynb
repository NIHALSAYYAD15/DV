Implementation of distribution-based graphs and charts Box, Heatmap.

### 1. Box 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data generation
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Box Plot')
plt.xlabel('Values')
plt.show()
```

### 2. Heatmap 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data generation (correlation matrix)
data_corr = pd.DataFrame(np.random.randn(10, 5), columns=[f'var_{i}' for i in range(5)])

# Heatmap (Correlation Matrix)
plt.figure(figsize=(8, 6))
sns.heatmap(data_corr.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap (Correlation Matrix)')
plt.show()
```
