Implementation of comparison - based graphs and charts â€“ Lollipop, Bullet.

### 1. Lollipop:

```python
import matplotlib.pyplot as plt

categories = ['Category 1', 'Category 2', 'Category 3']
values = [10, 15, 8]

plt.stem(categories, values, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Lollipop Chart Comparison')
plt.show()
```

### 2. Bullet:

```python
import matplotlib.pyplot as plt

categories = ['Category 1']
ranges = [5, 10, 15]  # Minimum, target, maximum values
actual = [12]

fig, ax = plt.subplots()
ax.set_aspect('equal')

ax.broken_barh([(ranges[0], ranges[1] - ranges[0]), (ranges[1], ranges[2] - ranges[1])], (0, 1), facecolors=('red', 'yellow'))
ax.broken_barh([(0, actual[0])], (0.4, 0.2), facecolors=('green'))

ax.set_yticks([0.5])
ax.set_yticklabels(categories)
ax.set_xticks(range(0, ranges[2] + 1, 5))
ax.set_xlabel('Values')

plt.title('Bullet Chart Comparison')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('Data/data2.csv')

# Extract the data for plotting
categories = data['Category'].tolist()
minimum = data['minimum'][0]
target = data['target'][0]
maximum = data['maximum'][0]
actual = data['Actual'][0]

# Plot the bullet chart
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Plot ranges
ax.broken_barh([(minimum, target - minimum), (target, maximum - target)], (0, 1), facecolors=('red', 'yellow'))
# Plot actual value
ax.broken_barh([(0, actual)], (0.4, 0.2), facecolors=('green'))

# Customize the chart
ax.set_yticks([0.5])
ax.set_yticklabels(categories)
ax.set_xticks(range(0, maximum + 1, 5))
ax.set_xlabel('Values')
plt.title('Bullet Chart Comparison')

# Show the chart
plt.show()
```
